```python
import pandas as pd
data=pd.read_csv('all.csv')
data['timestamp']= pd.to_datetime(data['timestamp'])
data=data.set_index(['timestamp'])
data.groupby(pd.Grouper(freq='D'))
Date=data.index
GMS=data['GMS'].to_list()
L_Date=[[Date[48*j+i] for i in range(48)] for j in range(383)]
L_GMS=[[GMS[48*j+i] for i in range(48)] for j in range(383)] 
Max_Date=[]
Max_GMS=[]
for i in range(len(L_Date)):
    for j in range(48):
        if L_GMS[i][j]==max(L_GMS[i]):
            Max_GMS.append(max(L_GMS[i]))
            Max_Date.append(L_Date[i][j])
A=[[Max_Date[i],Max_GMS[i]] for i in range(len(Max_Date))]
Data=pd.DataFrame(data=A, columns=['Time', 'Maximum GMS'])
Data.to_csv('Maximum_GMS.csv')
Data.head()
```
